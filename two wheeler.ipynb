```python
images=[]

for i in a:
   img=cv2.imread(i)
   images.append(img)
   cv2.imshow('image',img)
   cv2.waitKey()
   cv2.destroyAllWindows()
  

  
converted_img1=cv2.cvtColor(cv2.imread(img1),cv2.COLOR_BGR2RGB)
plt.imshow(converted_img1)






    
```

```python
face_cascade=cv2.CascadeClassifier(r'C:\Users\Basil Alias\Documents\Computer vision\Face Detection\haarcascade_frontalface_default.xml')
```

```python
face_cascade
```

```python
img_grey1=cv2.cvtColor(cv2.imread(img1),cv2.COLOR_BGR2GRAY)
face=face_cascade.detectMultiScale(img_grey1,scaleFactor=1.1,minNeighbors=4)
print(len(face))
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(converted_img1,(x,y),(x+w,y+h),color=(0,255,0),thickness=3)
plt.imshow(converted_img1)
if len(face)>2:
    print("Traffic rules violated")
```
